```python
from kafka import KafkaConsumer
import json
import pandas as pd

#  Step 1: Initialize Kafka Consumer
consumer = KafkaConsumer(
    'stock-prices',
    bootstrap_servers='localhost:9092',
    auto_offset_reset='earliest',
    enable_auto_commit=True,
    group_id='my-stock-consumer-group',
    value_deserializer=lambda x: json.loads(x.decode('utf-8'))
)

#  Step 2: Collect messages
messages = []

# Let's consume a limited number of messages (e.g., 100)
for idx, message in enumerate(consumer):
    messages.append(message.value)
    
    if idx >= 100:  # break after 100 messages
        break

#  Step 3: Convert to DataFrame
df = pd.DataFrame(messages)

#  Step 4: Display or export
print(df.head())           # preview in console
# df.to_csv("kafka_stock_prices.csv", index=False)  # optional export
```

```python
from kafka import KafkaConsumer
import pandas as pd
import json

# Kafka Consumer
consumer = KafkaConsumer(
    'stock-prices',
    bootstrap_servers='localhost:9092',
    auto_offset_reset='earliest',
    enable_auto_commit=True,
    group_id='stock-price-group',
    value_deserializer=lambda m: json.loads(m.decode('utf-8'))
)

# Collect messages
records = [msg.value for _, msg in zip(range(100), consumer)]

# Normalize nested JSON
df = pd.json_normalize(records)

# Display clean tabular structure
print(df.head(5))
```
